```python
import os 
import cv2 
import numpy as np 
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from sklearn.utils import shuffle
```

```python
def load_images_from_folder(folder, label):
    images = []
    labels = []
    for filename in os.listdir(folder):
        if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            img_path = os.path.join(folder, filename)
            image = cv2.imread(img_path)
            if image is not None:
                image = cv2.resize(image, (224, 224))   #resize ảnh về kích thước chuẩn
                image = img_to_array(image) #chuyển ảnh sang mảng số (tensor)
                image = preprocess_input(image) #chuẩn hóa theo mobilenetv2
                images.append(image)
                labels.append(label)    #gán nhãn: 1 cho stroke, 0 cho nostroke 
    print(filename)
    return images, labels 
```
